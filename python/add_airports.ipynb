```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', 30)
pd.set_option('display.max_rows', 4000)
```

```python
data = pd.read_csv("data/raw_data.csv", sep=",")
data.head(5)
```

```python
airports = pd.read_csv("data/airports.csv", sep=",")
airports = airports.rename(columns = {'iata':'depAirport'})
airports = airports.drop('country', 1)
airports.head(5)
```

```python
merged = pd.merge(data, airports, on = 'depAirport', how='inner')
merged.head(5)
```

```python
merged.to_csv('data/delaysAirports.csv', index=False, encoding='utf-8')
```
